## Preprocesamiento del dataset. Transformaciones previas necesarias para la modelación

```python
#Código que realice las transformaciones necesarias para poder realizar los procesos de modelación. Ej.One hot enconding
#el dataset elegido ya tiene las trasnformaciones pertinentes hechas.
df['Income'] = df['Income'].astype(int)
```

## División del dataset en datos de entrenamiento y datos de test 

```python
#Código que realice la división en entrenamiento y test, de acuerdo con la estretgia de evaluación planeada. Describa cuál es.
X = df[["Income", "Age", "MntWines", "NumDealsPurchases", "NumWebPurchases"
                 , "NumCatalogPurchases", "NumStorePurchases", "NumWebVisitsMonth"]].values
y = df[["MntTotal"]].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

## Propuesta de arquitectura de red neuronal

```python
print(X_train)
print(y_train)
```

Describe: 

+ las neuronas en la capa de entrada 
+ las capas intermedias – al menos dos – 
+ capa de salida
+ funciones de activación

```python
# Código de la estructuración de la red
epochs = 20
batch_size = 20


model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[8]))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(64, activation="relu"))
model.add(keras.layers.Dense(1))

```

```python
#Código de la inspección del modelo de red
model.summary()
```

##  Ajuste de modelo de Regresión RNA

Mediante Python y utilizando al menos Keras sobre TensorFlow 2.0 (tensorflow.keras), entrena el modelo o modelos de red neuronal escogidos.

```python
#Código de ajuste y entrenamiento
model.compile(loss='mse',
              optimizer='adam',
              metrics=['mse', 'mae'])

history = model.fit(X_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(X_test,y_test))
```

## Evaluación de modelo RNA

Defina las estadísticas (métricas) de evaluación, y dividiendo el dataset en datos de entrenamiento, validación y datos de test prueba tu propuesta.


### Visualice el progreso de entrenamiento del modelo y muestre las estadísticas de evaluación para los conjuntos de entrenamiento y validación.  

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()

#loss
training_loss = history.history['loss']
test_loss = history.history['val_loss']


epoch_count = range(1, len(training_loss) + 1)

# visualiza loss
plt.plot(epoch_count, training_loss, 'r--')
plt.plot(epoch_count, test_loss, 'b-')
plt.legend(['Training Loss', 'Test Loss'])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('training loss')
plt.show()


#mse
training_mse = history.history['mse']
test_mse = history.history['val_mse']


epoch_count = range(1, len(training_loss) + 1)

# Visualiza MSE
plt.plot(epoch_count, training_mse, 'r--')
plt.plot(epoch_count, test_mse, 'b-')
plt.legend(['Training MSE', 'Test MSE'])
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.title('training MSE')
plt.show()


test_predictions = model.predict(X_test).flatten()
a = plt.axes(aspect='equal')
plt.scatter(y_test, test_predictions)
plt.xlabel('X')
plt.ylabel('y')
_=plt.plot(y,y,color='orange')

plt.figure(figsize=(10,10))
plt.scatter(y_test, test_predictions, c='crimson')

p1 = max(max(test_predictions), max(y_test))
p2 = min(min(y_test), min(test_predictions))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

```python
#Código de evaluación de la red propuesta (entrenamiento y validación)
score = model.evaluate(X_train, y_train, verbose=1)
print('Train loss; ', score[0])
print('Train mse: ', score[1])

score = model.evaluate(X_test, y_test, verbose=1)
print('Test loss; ', score[0])
print('Test mse: ', score[1])
```

## Ajuste de modelos de Regresión alternativos

Elige al menos un método de regresión no basado en redes neuronales (p.ej. regresión lineal, regresión polinómica, regresión logarítmica, SVR, random forest regression, etc.).

```python
#Código de ajuste del modelo 1
model_RF = RandomForestRegressor(n_jobs=2, random_state=1, n_estimators=500)
model_RF.fit(X_train, y_train.ravel())
predictions = model_RF.predict(X_test)
print(mean_squared_error(y_test, predictions, squared=False))
```

```python
#Código para mostrar la evaluación de los modelos
# The coefficients
print("Coefficients: \n", model_RF.feature_importances_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, predictions))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, predictions))
```

```python
#Código de ajuste del modelo 2
model_LR = LinearRegression()
model_LR.fit(X_train, y_train.ravel())
predictions = model_LR.predict(X_test)
```

```python
#Código para mostrar la evaluación de los modelos
# The coefficients
print("Coefficients: \n", model_LR.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, predictions))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, predictions))
```
